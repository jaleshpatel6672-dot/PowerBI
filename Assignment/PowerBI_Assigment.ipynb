## 1) What is Power BI and how does it differ from Excel?


## Ans:- Microsoft Power BI is a business analytics and data visualization tool used to analyze large datasets and create interactive dashboards and reports.
## Microsoft Power BI focuses on advanced data visualization and dashboards, while Microsoft Excel is mainly used for data entry, calculations, and basic analysis

## 2) Explain the concept of data modeling in Power BI.


## Ans:- Data modeling means organizing and connecting different data tables so that you can analyze data easily and get accurate insights.
## Tables: Different datasets (e.g., Sales, Customers, Products)
## Relationships: Connecting tables using a common column (like Customer ID)
## Primary Key & Foreign Key: Columns used to link tables
## Data Model View: Where you visually connect tables in Power BI

## 3) What are the different types of connections available in Power BI?


## Ans:- 1) Import Connection, 2) DirectQuery Connection, 3) Live Connectionm

## 4) How do you handle data transformation in Power BI?


## Ans:- 1) Open Power Query, 2) Clean the Data, 3) Change Data Types, 4) Filter & Sort Data, 5) Split / Merge Columns, 6) Create New Columns, 7) Group & Aggregate Data, 8) Apply & Load

## 5) What is DAX (Data Analysis Expressions) and why is it important in Power BI?



## Ans:- DAX (Data Analysis Expressions) is a formula language used in Power BI to perform calculations and create custom logic on data.
## Create Measures, Advanced Calculations, Custom Logic, Dynamic Reports, Better Insights

## 6) Can you explain the difference between calculated columns and measures in Power BI?



## Ans:- | Point       | Calculated Column    | Measure           |
## | ----------- | -------------------- | ----------------- |
## | Calculation | Row-wise             | Aggregate         |
## | Storage     | Stored in model      | Not stored        |
## | Performance | Slower (uses memory) | Faster            |
## | Use         | Table data           | Visuals & reports |
## | Behavior    | Static               | Dynamic           |


## 7) How do you handle relationships between tables in Power BI?

## 8) What is the purpose of a Power BI Gateway?


## Ans:- A Power BI Gateway is a tool that connects your local (on-premises) data sources to Power BI Service (cloud) securely.
## Access Local Data, Automatic Data Refresh, Secure Connection, Real-Time Access

## 9) How can you schedule data refresh in Power BI Service?

## Ans:- 1.Publish Report, 2.Go to Workspace, 3.Open Settings, 4.Configure Data Source Credentials, 5.Set Up Gateway (if using local data), 6.Enable Scheduled Refresh, 7.Set Frequency, 8.Save Settings

## 10) Explain the concept of row-level security in Power BI.


## Ans:- Row-Level Security (RLS) is used to restrict data access for different users so they can only see the data that is relevant to them.

## 11) What is the Power BI Desktop and how does it differ from Power BI Service?


## Ans:- Power BI Desktop is used to create reports, while Power BI Service is used to publish, share, and manage them online.

## 12) Explain the concept of Direct Query in Power BI.

## Ans:- DirectQuery is a data connection mode where Power BI does NOT store the data internally, but instead fetches data directly from the source whenever you interact with the report.

## 13) What are Power BI templates and how are they useful?

## Ans:- Power BI Templates are pre-designed report files (.PBIT) that contain:
## Data model
## Report layout (charts, visuals)
## Queries and transformations
## USEFULL : Save Time, Reusability, Standardization, Easy Sharing
